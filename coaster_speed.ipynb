```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# set x-axis range
seconds = np.arange(0, 130, 10)
seconds
```

```python
# define datapoint variables
danger_speed = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
rail_speed = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
# plot points for both coasters
danger, = plt.plot(seconds, danger_speed, color="red", label="Danger Drop")
rail, = plt.plot(seconds, rail_speed, color="green", label="RailGun")
```

```python
# add title, axis labels, legend
plt.title("Coaster Speed Over Time")
plt.xlabel("Coaster Run Time")
plt.ylabel("Speed (MPH)")
plt.legend(loc="best")
```

```python
# set limits for plot to 0 and 120
plt.xlim(0, 120)
plt.ylim(5, 95)
```

```python
# add gridlines
plt.grid()
```
