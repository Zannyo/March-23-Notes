```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set x axis to numerical value for month
x_axis = np.arange(1,13,1)
x_axis
```

```python
# Avearge weather temp
Fdegrees = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Convert to Celsius C = (F-32) * 0.56
# THIS IS A LIST COMPREHENSION. WE TALKED ABOUT IT WITH ROGER,NOT SHAUN.
Cdegrees = [(x-32) * 0.56 for x in Fdegrees]
Cdegrees
```

```python
# create handles for each temperature list, pull in commands for markers
# if you don't define with the , it won't return the correct points because it needs to be named/formatted in a certain way
Fahrenheit, = plt.plot(x_axis, Fdegrees, marker ='+', color='blue', label="Fahreneit")
Celsius, = plt.plot(x_axis, Cdegrees, marker ='s', color='red', label="Celsius")
```

```python
# create legend
# can use this function to only plot certain handles if you designate, if you don't include "handle=[]", then it adds all variables to legend
plt.legend(loc="upper right")
```

```python
# change x and y axis labels
plt.xlabel("Months")
plt.ylabel("Degrees")
```
